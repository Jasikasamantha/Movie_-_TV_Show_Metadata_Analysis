```python
# import python libraries

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
import ast
```

```python
# import csv file
df = pd.read_csv(r'/Users/noja/Downloads/IMBD 2.csv')
```

```python
df.shape
```

```python
df.head(20)
```

```python
df.info()
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
# Clean runtime
df['runtime'] = df['runtime'].astype(str).str.extract('(\d+)').astype(float)
```

```python
df.head(20)
```

```python
# Clean votes
df['votes'] = df['votes'].astype(str).str.replace(',', '', regex=True)
df['votes'] = pd.to_numeric(df['votes'], errors='coerce')
```

```python
df.head(10)
```

```python
# Clean director and stars
df['director'] = df['director'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) and x.startswith('[') else [])
df['stars'] = df['stars'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) and x.startswith('[') else [])
```

```python
df.head(10)
```

```python
# Fill or drop missing
df['certificate'].fillna("Not Rated", inplace=True)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df.dropna(subset=['rating'], inplace=True)
```

```python
df.head(10)
```

```python
# Add a helper column for primary director/star
df['primary_director'] = df['director'].apply(lambda x: x[0] if x else np.nan)
df['primary_star'] = df['stars'].apply(lambda x: x[0].strip() if x else np.nan)
```

```python
df.head(10)
```

```python
df.shape
```

```python
# Distribution of Movie Ratings
sns.histplot(df['rating'], bins=15, kde=True, color='blue')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
#Insight:

# Most movie ratings seem to be clustered around the 6â8 range, suggesting that movies generally get average to good reviews.
```

```python
# Top 10 Most Voted Titles
top_voted = df.sort_values(by='votes', ascending=False).head(10)
sns.barplot(x='votes', y='movie', data=top_voted)
plt.title('Top 10 Most Voted Titles')
plt.xlabel('Votes')
plt.ylabel('Title')
plt.show()
```

```python
# Sample genres column
df['genre'] = df['genre'].fillna('')
all_genres = df['genre'].str.split(',').explode().str.strip()
genre_counts = all_genres.value_counts()
print(genre_counts)
```

```python
# Top Directors by Average Rating
top_directors = (
    df.groupby('primary_director')['rating']
    .mean()
    .sort_values(ascending=False)
    .head(30)
)

sns.barplot(x=top_directors.values, y=top_directors.index)
plt.title('Top 30 Directors by Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('Director')
plt.show()
```
