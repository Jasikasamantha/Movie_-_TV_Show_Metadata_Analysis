```python
# import python libraries

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
import ast
```

```python
# import csv file
df = pd.read_csv(r'/Users/noja/Downloads/IMBD 2.csv')
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
# Clean runtime
df['runtime'] = df['runtime'].astype(str).str.extract('(\d+)').astype(float)
```

```python
df.head(20)
```

```python
# Clean votes
df['votes'] = df['votes'].astype(str).str.replace(',', '', regex=True)
df['votes'] = pd.to_numeric(df['votes'], errors='coerce')
```

```python
df.head(10)
```

```python
# Clean director and stars
df['director'] = df['director'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) and x.startswith('[') else [])
df['stars'] = df['stars'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) and x.startswith('[') else [])
```

```python
df.head(10)
```

```python
# Fill or drop missing
df['certificate'].fillna("Not Rated", inplace=True)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df.dropna(subset=['rating'], inplace=True)
```

```python
df.head(10)
```

```python
# Add a helper column for primary director/star
df['primary_director'] = df['director'].apply(lambda x: x[0] if x else np.nan)
df['primary_star'] = df['stars'].apply(lambda x: x[0].strip() if x else np.nan)
```
